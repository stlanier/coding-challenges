## 1. [Two Sum ](https://leetcode.com/problems/two-sum/)
*Easy*

Given an array of integers `nums` and an integer `target`, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

```python
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        for num in nums:
            if target-num in nums[nums.index(num)+1:]:
                i1 = nums.index(num)
                i2 = nums.index(target-num, i1+1)
                return [i1, i2]
```

## 2. [Add Two Numbers](https://leetcode.com/problems/add-two-numbers/)

*Medium*

You are given two non-empty linked lists representing two non-negative integers. The digits are stored in reverse order, and each of their nodes contains a single digit. Add the two numbers and return the sum as a linked list.

You may assume the two numbers do not contain any leading zero, except the number 0 itself.

`Input: l1 = [2,4,3], l2 = [5,6,4]` <br>
`Output: [7,0,8]` <br>
Explanation: 342 + 465 = 807.

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def addTwoNumbers(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        val = (l1.val+l2.val)%10
        carry = (l1.val+l2.val)//10
        if (l1.next or l2.next or carry):
            l1 = l1.next if l1.next else ListNode();
            l2 = l2.next if l2.next else ListNode();
            l1.val+=carry
            next_step = self.addTwoNumbers(l1, l2)
        else:
            next_step = None
        return ListNode(val, next_step)
            
```

## 3. [Longest Substring Without Repeating Characters](https://leetcode.com/problems/longest-substring-without-repeating-characters/)
*Medium*

Given a string, `s`, find the length of the longest substring without repeating characters.

`Input: s = "abcabcbb"` <br>
`Output: 3` <br>
Explanation: The answer is "abc", with the length of 3.

**Constraints**

`0 <= s.length <= 5 * 10E4` <br>
`s` consists of English letters, digits, symbols, and spaces.

```python
class Solution(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        if (s == ''):
            return 0
        substrings = []
        for i in range(len(s)):
            loc = i
            substr = ''
            while (loc < len(s) and (not s[loc] in substr)):
                substr += s[loc]
                loc += 1
            substrings.append(substr)
        lens = [len(substr) for substr in substrings]
        return max(lens)
```

```python
## Better solution: sliding window with pointers

class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        # create a hash with 0 default value to keep a tab on repetitions 
        hash = {x:0 for x in s}
        mx = 0
        i = 0
        j = 0
        while(j<len(s)):
            # If the character is repeated in window then
            # 1. Update max length yet
            # 2. shrink the window from start pointer till there are no repititions
            # 3. do what you would normally do duraing expansion of window (increase end pointer and add character to hash)
            if hash[s[j]] > 0:
                mx = max(mx, j-i)
                while(s[i] != s[j]):
                    hash[s[i]] = 0
                    i += 1
                hash[s[i]] = 0
                i += 1
            hash[s[j]] = 1
            j += 1
        mx = max(mx, j-i)
        return mx 
```
